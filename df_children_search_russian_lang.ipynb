```python
df = pd.read_csv('data/Minjust2018.csv', 
                 header=None, 
                 usecols=[4, 27, 28, 33])
df.fillna('', inplace=True)
df.rename(columns={4: 'name', 
                   27: 'director', 
                   28: 'activity', 
                   33: 'founders'},
              inplace = True)
```
